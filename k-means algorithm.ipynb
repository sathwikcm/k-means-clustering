```python
class KMeansAlgo:
    import random
    
    def __init__(self):
        self.centroids = None
        self.k = None
        self.labels = None
    
    
    def train(self,X,num_clusters,num_iterations):
        
        self.k = num_clusters
        self.centroids = np.zeros([self.k,X.shape[1]])
        self.labels = np.zeros([X.shape[0]])
        
        for i in range(self.k):
            self.centroids[i] = X[random.randint(0,X.shape[0])]
            
        
        for each_iteration in range(num_iterations):
            
            for each_point in range(X.shape[0]):
                distances = []
                for each_centroid in range(self.k):
                    distances.append(np.linalg.norm(X[each_point]-self.centroids[each_centroid]))

                self.labels[each_point] = np.argmin(distances)


            for each_centroid in range(self.centroids.shape[0]):
                self.centroids[each_centroid] = np.mean(X[self.labels == each_centroid])
                
        
        print('Training is done')
        
        
    def predict(self, X):
        
        y_pred = np.zeros([X.shape[0]])
        
        for each_point in range(X.shape[0]):
                distances = []
                for each_centroid in range(self.k):
                    distances.append(np.linalg.norm(X[each_point]-self.centroids[each_centroid]))

                y_pred[each_point] = np.argmin(distances)
                
        return y_pred
        
          


                                     
k1 = KMeansAlgo()
k1.train(X,3,50)
y_pred1 = k1.predict(X)
y_pred1
```

```python
accuracy_score(y_true,y_pred1)
```
